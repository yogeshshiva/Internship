```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
team = []
for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)
team
```

```python
match = [] 
for i in soup.find('td',class_="rankings-block__banner--matches"):
    match.append(i.text)
match
```

```python
matche = []
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matche.append(i.text)
matches = matche[0:18:2]
matches
```

```python
add = match + matches
add
```

```python
point = []
for i in soup.find('td',class_="rankings-block__banner--points"):
    point.append(i.text)
point
```

```python
points = []
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    points.append(i.text)
pointss = points[1:19:2]
print(pointss)
```

```python
addd = point + pointss
addd
```

```python
rating = soup.find('td',class_="rankings-block__banner--rating u-text-right")
rating.text.split('\n                            ')[1]
```

```python
ratings = []
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
ratings
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Teams':team[0:10],'Matches':add[0:10],'Points':addd[0:10],'Rating':ratings[0:10]})
df
```
