```python
Brands = []
Product_Descriptions = []
Prices = []
```

```python
start = 0
end = 3
for page in range(start,end):
    brand_tag = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand_tag[0:100]:
        brands = i.text
        Brands.append(brands)
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```

```python
Brands
```

```python
start = 0
end = 3
for page in range(start,end):
    product_tag = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in product_tag[0:100]:
        product = i.text
        Product_Descriptions.append(product)
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```

```python
Product_Descriptions
```

```python
start = 0
end = 3
for page in range(start,end):
    price_tag = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in price_tag[0:100]:
        prices = i.text
        Prices.append(prices)
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```
