```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
soup
```

```python
restaurant = soup.find('a',class_="restnt-name ellipsis")
restaurant
```

```python
restaurant.text
```

```python
Restaurant = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Restaurant.append(i.text)
Restaurant
```

```python
location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
rating = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
```

```python
images = soup.find('img',class_="no-img")
images
```

```python
images.get('data-src')
```

```python
image = []
for i in soup.find_all('img',class_="no-img"):
    image.append(i.get('data-src'))
image
```

```python
cuisi = soup.find('span',class_="double-line-ellipsis")
cuisi
```

```python
cuisi.text.split('|')[1]
```

```python
cuisines = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisines.append(i.text.split('|')[1])
cuisines
```

```python
print(len(Restaurant),len(cuisines),len(location),len(rating),len(image))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Restaurant_Name':Restaurant,'Cuisine':cuisines,'Location':location,'Ratings':rating,'Image_URL':image})
df
```

```python
df['Image_URL']
```

```python
df.to_csv('DATA.csv')
```

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
headline = soup.find('a',class_="LatestNews-headline")
headline
```

```python
headline.text
```

```python
Headlines = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    Headlines.append(i.text)
Headlines
```

```python
time = soup.find('time',class_="LatestNews-timestamp")
time
```

```python
time.text
```

```python
Times = []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    Times.append(i.text)
Times
```

```python
link = soup.find('a',class_="LatestNews-headline")
link.get('href')
```

```python
Link = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    Link.append(i.get('href'))
Link
```

```python
print(len(Headlines),len(Times),len(Link))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Headline':Headlines,'Time':Times,'News_Links':Link})
df
```
