```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Name = soup.find('div',class_="presidentListing")
Name
```

```python
Name.text
```

```python
Name.text.split("\n")[1].split('(')[0]
```

```python
Name = []
for i in soup.find_all('div',class_="presidentListing"):
    Name.append(i.text.split('\n')[1].split('(')[0])
Name
```

```python
Term = soup.find('div',class_="presidentListing")
Term
```

```python
Term.text
```

```python
Term.text.split('\n')[2].split('Term of Office')[1].split(':')[1]
```

```python
Term_of_Office = []
for i in soup.find_all('div',class_="presidentListing"):
    Term_of_Office.append(i.text.split('\n')[2].split('Term of Office')[1].split(':')[1])
Term_of_Office
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({"Name":Name,"Term of Office":Term_of_Office})
data
```
